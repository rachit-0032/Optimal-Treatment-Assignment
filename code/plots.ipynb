```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
full_df = pd.read_csv('data/prostate.csv')
```

```python
# read ATE_sg_10_5.csv
df = pd.read_csv('ATE_sg_10_5.csv')
df.sort_values(by=['rx'], inplace=True, ascending=True)
```

```python
d = full_df.groupby("rx")["dtime"].mean()
labels = df["rx"].unique()
plt.bar(labels, d)

plt.axvline(x=2.5, color='black', linestyle='--')

plt.ylabel('Months of survival')
plt.title('Months of survival by treatment')
```

```python
import matplotlib.pyplot as plt
import numpy as np

labels = df["rx"].unique()
men_means = np.around(df[df["sg_10_5"] == 1]["dtime_mean"].values, 2)
women_means = np.around(df[df["sg_10_5"] == 0]["dtime_mean"].values, 2)

x = np.arange(len(labels))  # the label locations
width = 0.35  # the width of the bars

fig, ax = plt.subplots()
rects1 = ax.bar(x - width/2, men_means, width, label='High severity (sg > 10.5)')
rects2 = ax.bar(x + width/2, women_means, width, label='low severity (sg <= 10.5)')

# Add some text for labels, title and custom x-axis tick labels, etc.
ax.set_ylabel('Months of survival')
ax.set_title('Months of survival by cancer severity')
ax.set_xticks(x, labels)
ax.legend(loc='lower right')

ax.bar_label(rects1, padding=1)
ax.bar_label(rects2, padding=1)

#create a horizontal line with mean
# ax.axhline(y=men_means.mean(), color='blue', linestyle='--')
# ax.axhline(y=women_means.mean(), color='orange', linestyle='--')

# crate vertical line before the first treatment
ax.axvline(x=2.5, color='black', linestyle='--')

fig.tight_layout()

plt.show()
```
